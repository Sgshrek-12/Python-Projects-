```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the data from the text file into a DataFrame
df = pd.read_csv('yob2023.txt', header=None, names=['name', 'gender', 'birth_count'])

# Display the first 5 rows
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print(df.info())
```

```python
# Compute the number of male and female babies born and store the results in `gender_counts`
gender_counts = df.groupby('gender')['birth_count'].sum()

# Print the output `gender_counts`
print(gender_counts.to_markdown(numalign="left", stralign="left"))
```

```python
# Sort the DataFrame in descending order by `birth_count` and store it in `df_sorted`
df_sorted = df.sort_values(by='birth_count', ascending=False)

# Extract the first 100 rows from `df_sorted` and store it in `top_100`
top_100 = df_sorted.head(100)

# Group `top_100` by `name` and sort it in descending order by `birth_count` and store it in `popular_names`
popular_names = top_100.groupby('name')['birth_count'].sum().sort_values(ascending=False)

# Print the output `popular_names`
print(popular_names.to_markdown(numalign="left", stralign="left"))
```
